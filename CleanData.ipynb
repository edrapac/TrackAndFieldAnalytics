## 100m Cleanup

First we will clean up the mens 100m data, we drop some placeholder stuff, add stride rate, and then drop rows that have NA for splits

```python
import pandas as pd
```

```python
mens_100 = pd.read_csv('100M Data/CSVs/Mens_100m_by_athlete_formatted.csv')
mens_100 = mens_100.dropna(subset=['Athlete']) # Drop our placeholder rows
mens_100.head()
```

```python
mens_100['Ath_Mt_Strd']
```

```python
# First some data cleanup 
def striderate(row):
    if isinstance(row['Ath_Mt_Strd'],str):
        entry = row['Ath_Mt_Strd'].split("_")
        if len(entry) == 3:
            striderate = entry[2]
            return striderate
    else:
        return None

mens_100['Stride Rate'] = mens_100.apply(striderate,axis=1)
mens_100.drop('Ath_Mt_Strd',axis=1,inplace=True)
```

```python
mens_100 = mens_100.dropna(subset=['10 m']) # Seems like there are some performances missing splits
mens_100
```

```python
#Looks good lets save it
mens_100.to_csv('Mens_100m_formatted.csv')
```
